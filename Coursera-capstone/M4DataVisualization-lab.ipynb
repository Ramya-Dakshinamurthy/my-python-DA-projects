Connect to the database.

Import pandas module.


```python
import pandas as pd
```

## Demo: How to run an sql query


```python
# print how many rows are there in the table named 'master'
QUERY = """
SELECT COUNT(*)
FROM master
"""

# the read_sql_query runs the sql query and returns the data as a dataframe
df = pd.read_sql_query(QUERY,conn)
df.head()
```

## Demo: How to list all tables


```python
# print all the tables names in the database
QUERY = """
SELECT name as Table_Name FROM
sqlite_master WHERE
type = 'table'
"""
# the read_sql_query runs the sql query and returns the data as a dataframe
pd.read_sql_query(QUERY,conn)
```

## Demo: How to run a group by query


```python
QUERY = """
SELECT Age,COUNT(*) as count
FROM master
group by age
order by age
"""
pd.read_sql_query(QUERY,conn)
```

## Demo: How to describe a table


```python
table_name = 'master'  # the table you wish to describe

QUERY = """
SELECT sql FROM sqlite_master
WHERE name= '{}'
""".format(table_name)

df = pd.read_sql_query(QUERY,conn)
print(df.iat[0,0])
```

## Visualizing distribution of data

```python
import matplotlib.pyplot as plt
%matplotlib inline
import seaborn as sns
```

## Visualizing relationships in data


### Scatter Plots


Create a scatter plot of `Age` and `WorkWeekHrs.`


```python
# your code goes here
QUERY = """ SELECT * FROM Master"""
df = pd.read_sql_query(QUERY,conn)
plot = sns.scatterplot(x = 'Age', y ='WorkWeekHrs', data = df)
```

### Bubble Plots


Create a bubble plot of `WorkWeekHrs` and `CodeRevHrs`, use `Age` column as bubble size.


```python
# your code goes here
QUERY = """ SELECT WorkWeekHrs, CodeRevHrs, Age from MASTER"""
df = pd.read_sql_query(QUERY,conn)
sns.scatterplot(x='WorkWeekHrs', y='CodeRevHrs', data=df, size='Age', color='green',hue='Age', alpha=0.5,sizes=(10, 500))
plt.title('WorkWeekHrs and CodeRevHrs By Age', size = 10)
plt.xlabel('WorkWeekHrs')
plt.ylabel('CodeRevHrs')
plt.show()
```

## Visualizing composition of data


### Pie Charts


Create a pie chart of the top 5 databases that respondents wish to learn next year. Label the pie chart with database names. Display percentages of each database on the pie chart.


```python
# your code goes here
QUERY = """SELECT DatabaseDesireNextYear, COUNT(*) as count FROM DatabaseDesireNextYear group by DatabaseDesireNextYear
order by count(DatabaseDesireNextYear) desc limit """
df=pd.read_sql_query( QUERY, conn)
df.set_index('DatabaseDesireNextYear',inplace=True)
df['count'].plot(kind='pie', figsize=(20,6), autopct='%1.1f%%', labels=None, startangle=90,shadow=True, pctdistance=1.12)  
plt.title('Top 5 Databases Respondents Wish To Learn') 
plt.legend(df.index,loc='upper right')
plt.axis('equal')
plt.show()
```

```python
QUERY = """SELECT LanguageDesireNextYear, COUNT(*) as count FROM LanguageDesireNextYear group by LanguageDesireNextYear
order by count(LanguageDesireNextYear) desc limit 5"""
df3=pd.read_sql_query( QUERY, conn)
df3
df3.set_index('LanguageDesireNextYear',inplace=True)
df3['count'].plot(kind='pie', figsize=(20,6), autopct='%1.1f%%', labels=None, startangle=90,shadow=True, pctdistance=1.12)  
plt.title('Top 5 LanguageDesireNextYear') 
plt.legend(df3.index,loc='upper right')
plt.axis('equal')
plt.show()
```

```python
QUERY = """SELECT LanguageWorkedWith, COUNT(*) as count FROM LanguageWorkedWith group by LanguageWorkedWith
order by count(LanguageWorkedWith) desc limit 5"""
df4=pd.read_sql_query( QUERY, conn)
df4.set_index('LanguageWorkedWith',inplace=True)
df4['count'].plot(kind='pie', figsize=(20,6), autopct='%1.1f%%', labels=None, startangle=90,shadow=True, pctdistance=1.12)  
plt.title('Top 5 LanguageWorkedWith') 
plt.legend(df4.index,loc='upper right')
plt.axis('equal')
plt.show()
```

## Visualizing comparison of data


### Line Chart


Plot the median `ConvertedComp` for all ages from 45 to 60.


```python
# your code goes here
QUERY = """SELECT ConvertedComp, Age FROM Master WHERE Age BETWEEN 45 AND 60"""
df=pd.read_sql_query(QUERY,conn)
median=df.groupby('Age').median()
median.plot(kind='line', figsize=(15,6))
plt.title('Median ConvertedComp for Those Age 45 to 60')
plt.ylabel('ConvertedComp')
plt.show()
```

### Bar Chart


Create a horizontal bar chart using column `MainBranch.`


```python
# your code goes here
QUERY="""SELECT MainBranch, COUNT(MainBranch) as Count FROM Master GROUP BY MainBranch """
df=pd.read_sql_query(QUERY,conn)
df.set_index('MainBranch')
df.plot(kind='barh',color='Gold',figsize=(10,6))
plt.show()
```
